```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import girvan_newman
```

```python
G = nx.Graph()
```

```python
G.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('B', 'C'), ('B', 'D'),
    ('C', 'E'), ('E', 'F'), ('F', 'G'), ('F', 'H'),
    ('G', 'H'), ('H', 'I'), ('I', 'J')
])
```

```python
plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray', font_size=12, font_weight='bold')
plt.title('Original Friend Network')
plt.show()
```

```python
def girvan_newman_communities(G):
    communities = girvan_newman(G)
    first_iteration = tuple(sorted(c) for c in next(communities))
    return first_iteration

communities = girvan_newman_communities(G)
print(f'Communities detected: {communities}')
```

```python
color_map = []
for node in G:
    if node in communities[0]:
        color_map.append('lightgreen')
    else:
        color_map.append('lightcoral')
```

```python
plt.figure(figsize=(8,6))
nx.draw(G, node_color=color_map, with_labels=True, node_size=1500, edge_color='gray', font_size=12, font_weight='bold')
plt.title('Friend of Friends Network (Detected Communities)')
plt.show()
```
